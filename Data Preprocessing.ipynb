```python
def assign_sorted(row):
    ideology = row["V161126"]
    party = row["V161158x"]

    if pd.isna(ideology) or pd.isna(party):
        return np.nan
    elif ideology <= 3 and party <= 3:
        return 1  # 진보 + 민주당
    elif ideology >= 5 and party >= 5:
        return 1  # 보수 + 공화당
    else:
        return 0  # 정렬되지 않음

filtered_data["sorted_2016"] = filtered_data.apply(assign_sorted, axis=1)
```

```python
import numpy as np
def compute_aff_pol(row):
    party_id = row["V161158x"]
    dem_score = row["V201156"]  
    rep_score = row["V201157"] 

    if pd.isna(dem_score) or pd.isna(rep_score):
        return np.nan


    if party_id in [1, 2, 3]:    
        return abs(dem_score - rep_score)
    elif party_id in [5, 6, 7]:     
        return abs(rep_score - dem_score)
    elif party_id in [4]:   
        return abs(dem_score - rep_score)
    else:
        return np.nan

filtered_data["affective_polarization_2020"] = filtered_data.apply(compute_aff_pol, axis=1)
```

```python
filtered_data["age_group"] = filtered_data["V161267x"]


filtered_data["age_group"] = filtered_data["age_group"].astype("category")
```

```python
filtered_data['V161267x'] = filtered_data['V161267x'].str.extract('(^-?\d+)').astype(float)
```

```python
age_map = {
    1: "18-20",  2: "21-24", 3: "25-29",
    4: "30-34",  5: "35-39", 6: "40-44",
    7: "45-49",  8: "50-54", 9: "55-59",
    10: "60-64", 11: "65-69", 12: "70-74",
    13: "75+"
}


filtered_data["age_group_13"] = filtered_data["V161267x"].map(age_map)
filtered_data.loc[filtered_data["V161267x"] == -2, "age_group_13"] = np.nan
filtered_data = filtered_data.dropna(subset=["age_group_13"])
```

```python
def collapse_age(code):
    if code in [1,2,3]:
        return "18-29"
    elif code in [4,5,6]:
        return "30-44"
    elif code in [7,8,9]:
        return "45-59"
    elif code in [10,11,12,13]:
        return "60+"
    else:
        return np.nan  


filtered_data["age_group_4"] = filtered_data["V161267x"].apply(collapse_age)


age4_dummies = pd.get_dummies(
    filtered_data["age_group_4"],
    prefix="age",
    dummy_na=False
)
filtered_data = pd.concat([filtered_data, age4_dummies], axis=1)
```

```python
race_map = {
    1: "White",
    2: "Black",
    3: "Asian_PI",
    4: "Native_American",
    5: "Hispanic",
    6: "Other"
}


filtered_data["race_cat"] = filtered_data["V161310x"].map(race_map)


filtered_data.loc[filtered_data["V161310x"] == -2, "race_cat"] = np.nan
filtered_data = filtered_data.dropna(subset=["race_cat"])

filtered_data["race_collapsed"] = filtered_data["race_cat"].replace({
    "Asian_PI":      "Other",
    "Native_American": "Other",
    "Other":          "Other"
})

race_dummies = pd.get_dummies(
    filtered_data["race_collapsed"],
    prefix="race",
    dummy_na=False
)


filtered_data = pd.concat([filtered_data, race_dummies], axis=1)
```
