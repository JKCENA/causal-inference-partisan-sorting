```python
import numpy as np
import pandas as pd
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import KFold
from sklearn.utils import resample


pre_treat = ['age_group_4','race_collapsed','region','pol_interest_index',
             'participation','social_political_days','scaled_channel_count']
W_raw = df[pre_treat].reset_index(drop=True)
W = pd.get_dummies(W_raw, drop_first=True) 
T = df['sorted_2016'].reset_index(drop=True).values 
M = df['affective_polarization_2016'].reset_index(drop=True).values
Y = df['affective_polarization_2020'].reset_index(drop=True).values
n = len(df)


m_bar = M.mean()


kf = KFold(n_splits=5, shuffle=True, random_state=42)


Y_1_m = np.zeros(n)
Y_0_m = np.zeros(n)

for train_idx, test_idx in kf.split(df):
    X_train = np.column_stack([T[train_idx], M[train_idx], W.iloc[train_idx].values])
    y_train = Y[train_idx]
    model_y = GradientBoostingRegressor(random_state=1).fit(X_train, y_train)

    X_test_1m = np.column_stack([np.ones(len(test_idx)), np.full(len(test_idx), m_bar), W.iloc[test_idx].values])
    X_test_0m = np.column_stack([np.zeros(len(test_idx)), np.full(len(test_idx), m_bar), W.iloc[test_idx].values])

    Y_1_m[test_idx] = model_y.predict(X_test_1m)
    Y_0_m[test_idx] = model_y.predict(X_test_0m)


CDE_mbar = np.mean(Y_1_m - Y_0_m)
print("CDE at m_bar:", CDE_mbar)

def compute_cde_boot(i):
    idx = resample(np.arange(n), replace=True, n_samples=n)
    Xb = np.column_stack([T[idx], M[idx], W.iloc[idx].values])
    yb = Y[idx]
    model_yb = GradientBoostingRegressor(random_state=1).fit(Xb, yb)
    X_1m_b = np.column_stack([np.ones(n), np.full(n, m_bar), W.values])
    X_0m_b = np.column_stack([np.zeros(n), np.full(n, m_bar), W.values])
    y1b = model_yb.predict(X_1m_b)
    y0b = model_yb.predict(X_0m_b)
    return np.mean(y1b - y0b)

B = 1000  
boots = [compute_cde_boot(b) for b in range(B)]
ci_low, ci_high = np.percentile(boots, [2.5, 97.5])
print("Bootstrap 95% CI for CDE:", ci_low, ci_high)
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from joblib import Parallel, delayed
import multiprocessing


pre_treat = ['age_group_4','race_collapsed','region','pol_interest_index',
             'participation','social_political_days','scaled_channel_count']

treatment = "sorted_2016"
mediator = "affective_polarization_2016"
outcome = "affective_polarization_2020"
weight_col = None   


def _prepare_W(df, covariates):
    if not covariates:
        return pd.DataFrame(index=df.index)
    W_raw = df[covariates].copy()
    W = pd.get_dummies(W_raw, drop_first=True, dtype=int)
    W.index = df.index
    return W

def _fit_wls(y, X, weights=None):
    Xc = sm.add_constant(X, has_constant='add')
    if weights is None:
        model = sm.OLS(y, Xc)
    else:
        model = sm.WLS(y, Xc, weights=weights)
    return model.fit()

def sequential_gformula_point(df, treatment, mediator, outcome, covariates, weight_col=None, return_models=False):
    data = df.copy().reset_index(drop=True)

    uniq = set(data[treatment].dropna().unique())
    if not uniq.issubset({0,1}):
        raise ValueError("treatment must be binary {0,1}")
    w = data[weight_col].values if (weight_col is not None and weight_col in data.columns) else None

    W = _prepare_W(data, covariates)
    A = data[treatment].values
    M = data[mediator].values
    Y = data[outcome].values


    X_M = np.column_stack([A.reshape(-1,1), W.values]) if W.shape[1] > 0 else A.reshape(-1,1)
    fit_M = _fit_wls(M, X_M, weights=w)


    X_M_A1 = np.column_stack([np.ones(len(A)).reshape(-1,1), W.values]) if W.shape[1] > 0 else np.ones(len(A)).reshape(-1,1)
    X_M_A0 = np.column_stack([np.zeros(len(A)).reshape(-1,1), W.values]) if W.shape[1] > 0 else np.zeros(len(A)).reshape(-1,1)
    mu_M_A1 = fit_M.predict(sm.add_constant(X_M_A1, has_constant='add'))
    mu_M_A0 = fit_M.predict(sm.add_constant(X_M_A0, has_constant='add'))


    X_Y = np.column_stack([A.reshape(-1,1), M.reshape(-1,1), W.values]) if W.shape[1] > 0 else np.column_stack([A.reshape(-1,1), M.reshape(-1,1)])
    fit_Y = _fit_wls(Y, X_Y, weights=w)

    def predY_vec(A_val, M_vec):
        Xpred = np.column_stack([np.full(len(A), A_val).reshape(-1,1), M_vec.reshape(-1,1), W.values]) if W.shape[1] > 0 else np.column_stack([np.full(len(A), A_val).reshape(-1,1), M_vec.reshape(-1,1)])
        return fit_Y.predict(sm.add_constant(Xpred, has_constant='add'))

    Y_1_M1 = predY_vec(1, mu_M_A1)
    Y_1_M0 = predY_vec(1, mu_M_A0)
    Y_0_M0 = predY_vec(0, mu_M_A0)

    TE = float(np.mean(Y_1_M1 - Y_0_M0))
    NDE = float(np.mean(Y_1_M0 - Y_0_M0))
    NIE = float(np.mean(Y_1_M1 - Y_1_M0))
    prop_med = float(NIE/TE) if abs(TE) > np.finfo(float).eps else np.nan

    out = {"TE":TE, "NDE":NDE, "NIE":NIE, "PropMediated":prop_med,
           "mu_M_A1":mu_M_A1, "mu_M_A0":mu_M_A0,
           "Y_1_M1":Y_1_M1, "Y_1_M0":Y_1_M0, "Y_0_M0":Y_0_M0}
    if return_models:
        out["fit_M"] = fit_M
        out["fit_Y"] = fit_Y
        out["W_columns"] = W.columns.tolist()
    return out

def interaction_test(df, treatment, mediator, outcome, covariates, weight_col=None):
    data = df.copy().reset_index(drop=True)
    W = _prepare_W(data, covariates)
    w = data[weight_col].values if (weight_col is not None and weight_col in data.columns) else None

    A = data[treatment].values
    M = data[mediator].values
    AM = (A * M).reshape(-1,1)
    X_int = np.column_stack([A.reshape(-1,1), M.reshape(-1,1), AM, W.values]) if W.shape[1] > 0 else np.column_stack([A.reshape(-1,1), M.reshape(-1,1), AM])
    fit_int = _fit_wls(data[outcome].values, X_int, weights=w)

    try:
        pval_am = float(fit_int.pvalues[3])
    except Exception:
        pval_am = np.nan
    return pval_am, fit_int.params


def _boot_worker_indices(data, idx, treatment, mediator, outcome, covariates, weight_col):
    d = data.iloc[idx].reset_index(drop=True)
    res = sequential_gformula_point(d, treatment, mediator, outcome, covariates, weight_col, return_models=False)
    return np.array([res["TE"], res["NDE"], res["NIE"], res["PropMediated"]], dtype=float)

def bootstrap_sequential_gformula(df, treatment, mediator, outcome, covariates, weight_col=None, R=1000, n_jobs=1, random_state=None):
    rng = np.random.RandomState(random_state)
    n = len(df)
    draws = [rng.choice(n, size=n, replace=True) for _ in range(R)]
    if n_jobs == -1:
        n_jobs = multiprocessing.cpu_count()
    results = Parallel(n_jobs=n_jobs)(
        delayed(_boot_worker_indices)(df, idx, treatment, mediator, outcome, covariates, weight_col)
        for idx in draws
    )
    boot_mat = np.vstack(results)
    names = ["TE","NDE","NIE","PropMediated"]
    point = sequential_gformula_point(df, treatment, mediator, outcome, covariates, weight_col, return_models=False)
    point_vec = np.array([point["TE"], point["NDE"], point["NIE"], point["PropMediated"]], dtype=float)
    ci_lower = np.percentile(boot_mat, 2.5, axis=0)
    ci_upper = np.percentile(boot_mat, 97.5, axis=0)
    summary = {"point": dict(zip(names, point_vec.tolist())),
               "boot_mean": dict(zip(names, np.mean(boot_mat, axis=0).tolist())),
               "boot_sd": dict(zip(names, np.std(boot_mat, axis=0, ddof=1).tolist())),
               "ci_2.5": dict(zip(names, ci_lower.tolist())),
               "ci_97.5": dict(zip(names, ci_upper.tolist())),
               "boot_mat": boot_mat}
    return summary


p_inter, params_int = interaction_test(df, treatment, mediator, outcome, pre_treat, weight_col)
print(f"A*M interaction p-value (outcome model): {p_inter:.4f}")
if (not np.isnan(p_inter)) and p_inter < 0.05:
    print("Warning: A*M interaction appears significant. Deterministic substitution (mu_M) may be invalid.")


res = sequential_gformula_point(df, treatment, mediator, outcome, pre_treat, weight_col, return_models=True)
print("\nPoint estimates:")
print(f"TE  = {res['TE']:.4f}")
print(f"NDE = {res['NDE']:.4f}")
print(f"NIE = {res['NIE']:.4f}")
print(f"Prop mediated = {res['PropMediated']:.3f}")

R = 1000  
n_jobs = 4  
boot_res = bootstrap_sequential_gformula(df, treatment, mediator, outcome, pre_treat, weight_col, R=R, n_jobs=n_jobs, random_state=2027)
print("\nBootstrap (percentile) 95% CIs:")
print(f"TE  : {boot_res['ci_2.5']['TE']:.3f} ~ {boot_res['ci_97.5']['TE']:.3f}")
print(f"NDE : {boot_res['ci_2.5']['NDE']:.3f} ~ {boot_res['ci_97.5']['NDE']:.3f}")
print(f"NIE : {boot_res['ci_2.5']['NIE']:.3f} ~ {boot_res['ci_97.5']['NIE']:.3f}")
print(f"Prop mediated CI: {boot_res['ci_2.5']['PropMediated']:.3f} ~ {boot_res['ci_97.5']['PropMediated']:.3f}")
```

```python
p_treated = y.mean()
p_control = 1 - p_treated
df['w_stab'] = np.where(
    df["sorted_2016"] == 1, p_treated / df["ps"], p_control / (1 - df["ps"])
)


plt.hist(df['w_stab'], bins=30)
plt.xlabel('Stabilized IP Weight')
plt.ylabel('Count')
plt.title('Distribution of Stabilized IP Weights')
plt.show()

# 4. Weighted SMD calculation
def weighted_mean(x, w):
    return np.sum(x * w) / np.sum(w)

def weighted_var(x, w):
    mu = weighted_mean(x, w)
    return np.sum(w * (x - mu)**2) / np.sum(w)

def compute_weighted_smd(x1, w1, x0, w0):
    m1, m0 = weighted_mean(x1, w1), weighted_mean(x0, w0)
    s1, s0 = np.sqrt(weighted_var(x1, w1)), np.sqrt(weighted_var(x0, w0))
    s_pooled = np.sqrt((s1**2 + s0**2)/2)
    return (m1 - m0) / s_pooled

treat_idx = df["sorted_2016"] == 1
control_idx = df["sorted_2016"] == 0
smd_dict = {}


for c in X.columns:
    smd = compute_weighted_smd(
        X.loc[treat_idx, c], df.loc[treat_idx, 'w_stab'],
        X.loc[control_idx, c], df.loc[control_idx, 'w_stab']
    )
    smd_dict[c] = smd

smd_df = pd.DataFrame(list(smd_dict.items()), columns=["Variable", "SMD"]).sort_values(by="SMD", key=np.abs, ascending=False)
print(smd_df)

# 5. Love plot
plt.figure(figsize=(6, min(12, len(smd_df)/2)))
plt.hlines(y=smd_df["Variable"], xmin=0, xmax=smd_df["SMD"], color='skyblue')
plt.plot(smd_df["SMD"], smd_df["Variable"], "o")
plt.axvline(0.1, color='red', linestyle='--', label='|SMD|=0.1')
plt.axvline(-0.1, color='red', linestyle='--')
plt.xlabel('Weighted Standardized Mean Difference (SMD)')
plt.ylabel('Covariate (one-hot)')
plt.title('Love Plot (Weighted Covariate Balance)')
plt.legend()
plt.tight_layout()
plt.show()

```
